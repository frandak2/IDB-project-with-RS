```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from matplotlib import pyplot as plt
# Say, "the default sans-serif font is COMIC SANS"
plt.rcParams['font.sans-serif'] = "Times New Roman"
plt.rcParams.update({'font.size': 20,"font.weight":"bold"})
```

```python
def autolabel(bar):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for p in bar.patches:
        bar.annotate(format(p.get_height(), '.2f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center', xytext = (0, 10), 
                     textcoords = 'offset points')
```

```python
dataset = pd.read_csv("acc.csv",sep=",")
rf = dataset[dataset.Method=="RF"]
svm = dataset[dataset.Method=="SVM"]
# dataset
```

```python

sns.set(style="white", context="talk")
plt.rcParams["font.family"] = "Times New Roman"
# Set up the matplotlib figure
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 12))


bar = sns.barplot(x="Features", y="Accuracy", hue="Sensor", palette="deep", ax=ax1, data=rf)
autolabel(bar)
ax1.set_xlabel("RF",fontweight='bold',fontsize=20)
ax1.set_ylabel("Accuracy",fontweight='bold',fontsize=20)
ax1.set(ylim=(0, 1.4))


bar = sns.barplot(x="Features", y="Accuracy", hue="Sensor", palette="deep", ax=ax2, data=svm)
autolabel(bar)
ax2.set_xlabel("SVM",fontweight='bold',fontsize=20)
ax2.set_ylabel("Accuracy",fontweight='bold',fontsize=20)
ax2.set(ylim=(0, 1.4))

plt.savefig("ACC.png",dpi=600)
```
