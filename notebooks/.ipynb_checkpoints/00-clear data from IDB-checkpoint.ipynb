```python
bs = pd.read_csv('Data_IDB/Panel data PATCA-DR - only baseline.csv', encoding='cp1252')
```

```python
names = ['cuest_id_b', 'parcela', 'cultivo','temporary_crop', 'superf_parcela_has', 'Y', 'X', 'patca']

# making new data frame with dropped NA values 
new_data = bs[names]
new_data
```

```python
cords_repY = new_data["Y"].value_counts()
# cords_repX = new_data["X"].value_counts()
```

```python
Y = cords_repY[cords_repY>1]
# X = cords_repX[cords_repX>1]
Y = Y.index
```

```python
cords_repY
```

```python
df = new_data
for i in Y:
    pos = new_data[new_data["Y"] == i].index
    pos = pos.delete(0)
    df = df.drop(pos)
```

```python
df["cultivo"][df["cultivo"].isnull()]="rest field"
df = df.dropna()

# df
```

```python
crs = {'init': 'epsg:4326'}
gdf = gpd.GeoDataFrame(
    df, crs=crs , geometry=gpd.points_from_xy(df.X, df.Y))
gdf.to_file(driver = 'ESRI Shapefile', filename= "Data_IDB/new/new_bl1.shp")
```
