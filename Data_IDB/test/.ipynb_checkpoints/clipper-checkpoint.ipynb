```python
import os, glob
import rasterio
from rasterio.plot import show
from rasterio.plot import show_hist
from rasterio.mask import mask
from shapely.geometry import box
import geopandas as gpd
from fiona.crs import from_epsg
import json
```

```python
g = os.getcwd()
g
path = 'D:\\OneDrive-CGIAR\\Workspace\\bank'
os.chdir(path)
```

```python
shp = gpd.read_file("Data_IDB/test/shp_test.shp")
```

```python
fp = 'Data_IDB/test/INERA.tif'
# open the lidar chm
data = rasterio.open(fp)
data
```

```python
show((data, 4), cmap='terrain')
```

```python
def getFeatures(gdf,ID):
    """Function to parse features from GeoDataFrame in such a manner that rasterio wants them"""
    import json
    return [json.loads(gdf.to_json())['features'][ID]['geometry']]
```

```python
coords = getFeatures(shp,2)
```

```python
out_img, out_transform = mask(data, shapes=coords, crop=True)
```

```python
out_meta = data.meta.copy()
```

```python
out_meta.update({"driver": "GTiff",
                "height": out_img.shape[1],
                "width": out_img.shape[2],
                "transform": out_transform})
```

```python
out_tif = 'Data_IDB/test/clip.tif'
with rasterio.open(out_tif, "w", **out_meta) as dest:
    dest.write(out_img)
```

```python
clipped = rasterio.open(out_tif)

show((clipped, 5), cmap='terrain')
```
